# Rhythm Frequencies of Pro Quinctio

```python
import sys
sys.path.append('/Users/tyler/Projects/latin-prose-rhythm')

from prose_rhythm.preprocessor import Preprocessor
from prose_rhythm.analyze import Analyze
```

```python
pro_quinctio_path = '/Users/tyler/Datasets/phi-macronized/cicero/pro_quinctio.txt'
with open(pro_quinctio_path) as f:
    text = f.read()

preprocess = Preprocessor(text=text)
analyze = Analyze()
```

```python
tokens = preprocess.tokenize()
rhythms = analyze.get_rhythms(tokens, include_sentence=False)
rhythm_frequencies = analyze.rhythm_frequency(rhythms)
```

```python
def print_frequencies(title, rhythm_frequencies):
    print(title)
    print('='*20)
    print('Total clausulae: {}'.format(rhythm_frequencies['total_clausulae']))
    print('Excluded clausulae: {}'.format(rhythm_frequencies['excluded_clausulae']))
    print('='*20)
    for k, v in rhythm_frequencies.items():
        if k is not 'total_clausulae' and k is not 'excluded_clausulae' and k is not 'x':
            print('{} : {}'.format(k, v))
```

```python
print_frequencies('Pro Quinctio', rhythm_frequencies)
```
