```python
import sys
sys.path.append('/Users/tyler/Projects/latin-prose-rhythm')

from prose_rhythm.preprocessor import Preprocessor
from prose_rhythm.analyze import Analyze
```

```python
import os
import re
import codecs

analyze = Analyze()

cicero_texts_path = '/Users/tyler/Datasets/phi-macronized/cicero/'
text_paths = [cicero_texts_path+p for p in os.listdir(cicero_texts_path)]

def print_frequencies(title, rhythm_frequencies):
    print(title)
    print('='*20)
    print('Total clausulae: {}'.format(rhythm_frequencies['total_clausulae']))
    print('Excluded clausulae: {}'.format(rhythm_frequencies['excluded_clausulae']))
    print('='*20)
    for k, v in rhythm_frequencies.items():
        if k is not 'total_clausulae' and k is not 'excluded_clausulae' and k is not 'x':
            print('{} : {}'.format(k, v))
```

```python
for path in text_paths:
    with codecs.open(path, encoding='utf-8', errors='ignore') as f:
        text = f.read()
    try:
        title = ' '.join([w.title() for w in path.split('/')[-1][:-4].split('_')])
        tokens = Preprocessor(text=text).tokenize()
        rhythms = analyze.get_rhythms(tokens, include_sentence=False)
        frequencies = analyze.rhythm_frequency(rhythms)
        print_frequencies(title, frequencies)
        print('\n'*3)
    except:
        print(title + ' problem with text')
        print('\n'*3)
```
