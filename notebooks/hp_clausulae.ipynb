```python
import os
import sys
module_path = os.path.abspath(os.path.join('..'))
if module_path not in sys.path:
    sys.path.append(module_path)
from prose_rhythm.preprocessor import Preprocessor
from prose_rhythm.analyze import Analyze
```

```python
with open('/Users/tyler/Desktop/hp/latin_hp.txt', encoding='utf-8') as f:
    text = f.read()
```

```python
tokens = Preprocessor(text=text, title='Harry Potter').tokenize()
```

```python
analysis = Analyze()
rhythms = analysis.get_rhythms(tokens)
```

```python
for i, sent in enumerate(rhythms):
    if type(sent) != int and sent[0].strip():
        print('Sentence {}: {}'.format(i, sent[0]))
        print('Rhythm: {}'.format(sent[1]))
```
